```python
from transformers import pipeline
```

```python
# # Load Summarization Model
# summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """
Generative Artificial Intelligence refers to models capable of producing new content such as text, images, audio, and code.
These systems learn patterns from large datasets and generate outputs that resemble human-created content.
Applications include chatbots, content creation, summarization, and code generation.
However, generative AI also introduces risks such as hallucinations, bias, and deepfakes,
requiring careful monitoring and ethical considerations.
"""
```

```python
# # generate summary

summary = summarizer(text, max_length=50, min_length=10, do_sample=False)
print(summary[0]['summary_text'])

```
